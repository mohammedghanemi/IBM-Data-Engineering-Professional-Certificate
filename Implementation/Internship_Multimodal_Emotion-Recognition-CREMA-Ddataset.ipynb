```python
import os
import pandas as pd

# Specify the path to the CSV file
csv_file_path = r'D:\CREMA-D\\VideoDemographics.csv'

# Read the CSV file into a pandas DataFrame
df_csv = pd.read_csv(csv_file_path)

# Create an empty list to store combined information
combined_data = []

# Specify the path to the folder containing .flv files
flv_folder_path = r'D:\CREMA-D\mp4_files'

# Iterate over .flv files
for file_name in os.listdir(flv_folder_path):
    if file_name.endswith('.mp4'):
        # Parse the file name to extract actor ID and emotion
        parts = file_name.split('_')
        actor_id = int(parts[0])
        emotion = parts[2]
        
        # Find corresponding information from CSV data
        actor_info = df_csv[df_csv['ActorID'] == actor_id]
        if not actor_info.empty:
            # Combine information
            combined_info = {
                'ActorID': actor_id,
                'Age': actor_info['Age'].values[0],
                'Sex': actor_info['Sex'].values[0],
                'Race': actor_info['Race'].values[0],
                'Ethnicity': actor_info['Ethnicity'].values[0],
                'Emotion': emotion
            }
            # Append combined information to list
            combined_data.append(combined_info)

# Convert the list of dictionaries to a DataFrame
combined_df = pd.DataFrame(combined_data)

# Display combined data
print(combined_df)

```


# Balancing The Data

```python
import os
import pandas as pd
from sklearn.utils import resample



# Create a multi-index based on Age, Sex, and Emotion
df_grouped = combined_df.groupby(['Age', 'Sex', 'Emotion'])

# Determine the maximum size of any group
max_size = df_grouped.size().max()

# Resample each group to the target size
balanced_groups = []

for name, group in df_grouped:
    if len(group) < max_size:
        group_resampled = resample(group,
                                   replace=True,  # Sample with replacement
                                   n_samples=max_size,  # To match the target size
                                   random_state=42)  # For reproducibility
    else:
        group_resampled = group
    balanced_groups.append(group_resampled)

# Combine all resampled groups back into a single DataFrame
df_balanced = pd.concat(balanced_groups)

# Shuffle the combined DataFrame to mix the resampled data
df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

# Verify the new distribution
print("Balanced DataFrame:")
print(df_balanced.groupby(['Age', 'Sex', 'Emotion']).size())

# Save the balanced DataFrame to a new CSV file if needed
df_balanced.to_csv('balanced_data.csv', index=False)
```

# Convert audio from video to Mel Spectrograms


```python
import os
import numpy as np
import librosa
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image

def save_mel_spectrogram(audio_path, output_image_path):
    y, sr = librosa.load(audio_path, sr=None)
    S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
    S_dB = librosa.power_to_db(S, ref=np.max)

    plt.figure(figsize=(10, 4))
    librosa.display.specshow(S_dB, sr=sr, x_axis='time', y_axis='mel')
    plt.colorbar(format='%+2.0f dB')
    plt.savefig(output_image_path)
    plt.close()


```

```python
def extract_spectrogram_features_batch(audio_directory, output_directory, start_index=0, batch_size=32):
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    audio_files = os.listdir(audio_directory)
    audio_files = audio_files[start_index:]  

    for i in range(3):  
        batch_files = audio_files[i * batch_size:(i + 1) * batch_size]

        for audio_file in batch_files:
            if audio_file.endswith(".mp3"):
                audio_path = os.path.join(audio_directory, audio_file)
                output_image_path = os.path.join(output_directory, f"{os.path.splitext(audio_file)[0]}.png")

                save_mel_spectrogram(audio_path, output_image_path)

```

```python
import os
from PIL import Image

def check_and_convert_all_images(directory_path, target_path):
    if not os.path.exists(target_path):
        os.makedirs(target_path)
    
    for filename in os.listdir(directory_path):
        if filename.endswith(".png") or filename.endswith(".jpg"):
            image_path = os.path.join(directory_path, filename)
            try:
                with Image.open(image_path) as img:
                    # Get image dimensions and mode
                    width, height = img.size
                    mode = img.mode
                    
                    # Convert to RGB if needed and resize
                    if mode == 'RGBA' or mode == 'L' or mode == 'P':
                        img = img.convert('RGB')
                    
                    # Resize to 224x224
                    img = img.resize((224, 224))

                    # Save the image to the new directory with RGB format
                    new_image_path = os.path.join(target_path, filename)
                    img.save(new_image_path)
                    
                    # Print the status
                    print(f"Processed {filename}: {width}x{height}, {mode} -> {img.size[0]}x{img.size[1]}, {img.mode}")

            except Exception as e:
                print(f"Error processing image {image_path}: {str(e)}")

# Example usage:
#directory_path = "D:/CREMA-D/Mel_Spectrograms"
#target_path = "D:/CREMA-D/MelSpectrograms_RGB"

#check_and_convert_all_images(directory_path, target_path)
```

```python
import os
import numpy as np
from keras.preprocessing import image
from keras.applications.vgg19 import VGG19, preprocess_input, decode_predictions
from keras.models import Model

# Function to load images, extract features using VGG19, and save them
def extract_features_and_save(directory_path, target_path):
    # Ensure the target directory exists or create it
    if not os.path.exists(target_path):
        os.makedirs(target_path)
    
    # Load the VGG19 model with pre-trained weights
    base_model = VGG19(weights='imagenet')
    # Remove the last fully connected layers to get the feature extractor
    model = Model(inputs=base_model.input, outputs=base_model.get_layer('fc2').output)
    
    # Iterate over the files in the directory
    for filename in os.listdir(directory_path):
        if filename.endswith(".png") or filename.endswith(".jpg"):
            image_path = os.path.join(directory_path, filename)
            try:
                # Load and preprocess the image
                img = image.load_img(image_path, target_size=(224, 224))
                img_array = image.img_to_array(img)
                img_array = np.expand_dims(img_array, axis=0)
                img_array = preprocess_input(img_array)
                
                # Extract features
                features = model.predict(img_array)
                
                # Save features to a .npy file
                feature_filename = os.path.splitext(filename)[0] + '.npy'
                feature_path = os.path.join(target_path, feature_filename)
                np.save(feature_path, features)
                
                # Get predictions from the original VGG19 model
                preds = base_model.predict(img_array)
                decoded_preds = decode_predictions(preds, top=3)[0]
                
                # Print the predictions
                print(f"Predictions for {filename}:")
                for i, (imagenet_id, label, score) in enumerate(decoded_preds):
                    print(f"    {i+1}: {label} ({score:.4f})")
                    
            except Exception as e:
                print(f"Error processing image {image_path}: {str(e)}")

# Example usage:
directory_path = "D:/CREMA-D/MelSpectrograms_RGB"
target_path = "D:/CREMA-D/Ftr_Extracted_MelRGB"

extract_features_and_save(directory_path, target_path)
print("Feature extraction and saving completed.")
```
